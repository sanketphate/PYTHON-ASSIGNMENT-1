```python
"""Q1). Key Features of Python That Make It a Popular Choice for Programming ?

Python has become one of the most widely used programming languages due to its simplicity, versatility, and powerful capabilities. Here are the key features that make it popular

1. Simple and Easy to Learn:
Python has a clean and readable syntax similar to English, making it easy for beginners to learn and use.

2. Interpreted Language:
Python code is executed line by line, which helps with debugging and makes the development process faster.

3. Dynamically Typed:
No need to declare the type of variables. Python automatically detects the type at runtime.

4. Extensive Standard Library:
Comes with a large number of built-in modules and functions, reducing the need to write code from scratch.

5. Cross-Platform Compatibility:
Python is platform-independent, meaning code written on one system can run on others with little or no modification.

6. Object-Oriented and Functional:
Supports both object-oriented programming (OOP) and functional programming paradigms.

7. Open Source and Community Support:
Python is free to use and has a large, active community that contributes libraries and tools.

8. Versatility:
Used in a wide range of applications including web development, data science, artificial intelligence, automation, and more."""








"""Q2). Role of Predefined Keywords in Python (with Examples):?

Predefined keywords in Python are reserved words that have special meanings and purposes in the language. These keywords are used to define the structure and syntax of Python programs and cannot be used as variable names.

Role of Keywords:

Control the flow of the program (e.g., if, else, while)
Define functions and classes (def, class)
Handle exceptions (try, except)
Declare values (True, False, None)
Manage variable scope (global, nonlocal)


Examples of Keywords in Use:"""

# Example 1: if-else keyword
x = 10
if x > 5:
    print("x is greater than 5")
else:
    print("x is 5 or less")

# Example 2: def keyword
def greet(name):
    return "Hello " + name

print(greet("Alice"))

# Example 3: while keyword
i = 1
while i <= 3:
    print(i)
    i += 1

"""List of Some Common Python Keywords:

if, else, elif, while, for, def, return, try, except, class, import, in, is, not, and, or, True, False, None"""











"""Q3). Compare and Contrast Mutable and Immutable Objects in Python (with Examples)?

In Python, objects are categorized as mutable or immutable based on whether their value can be changed after creation.

Mutable Objects:
Can be changed after creation.

Operations that modify the object do not change its identity (memory address stays the same).
Examples: list, dict, set"""


Example:

my_list = [1, 2, 3]
print(id(my_list))  # Before modification
my_list.append(4)
print(my_list)      # [1, 2, 3, 4]
print(id(my_list))  # Same ID, list was modified in place

"""Immutable Objects:

Cannot be changed after creation.

Any change creates a new object with a different identity.

Examples: int, float, str, tuple"""


Example:

x = "hello"
print(id(x))        # Before change
x = x + " world"
print(x)            # "hello world"
print(id(x))        # New ID, new string object created









"""Q4). Discuss the different types of operators in Python and provide examples of how they are used ?
1. Arithmetic Operators: These operators perform basic mathematical calculations.
+ (Addition): a + b (e.g., 2 + 3 results in 5).
- (Subtraction): a - b (e.g., 5 - 2 results in 3).
* (Multiplication): a * b (e.g., 2 * 3 results in 6).
/ (Division): a / b (e.g., 10 / 3 results in 3.333...).
// (Floor Division): a // b (e.g., 10 // 3 results in 3 - returns the integer part of the quotient).
% (Modulo): a % b (e.g., 10 % 3 results in 1 - returns the remainder of the division).
`: ` (Exponentiation):** a ** b (e.g., 2 ** 3 results in 8 - raises a to the power of b).


2. Comparison Operators: These operators compare two values and return a boolean result (True or False).
> (Greater than): a > b (e.g., 5 > 2 results in True).
>= (Greater than or equal to): a >= b (e.g., 5 >= 5 results in True).
< (Less than): a < b (e.g., 2 < 5 results in True).
<= (Less than or equal to): a <= b (e.g., 5 <= 5 results in True).
== (Equal to): a == b (e.g., 5 == 5 results in True).
!= (Not equal to): a != b (e.g., 5 != 2 results in True).


3. Logical Operators: These operators are used to combine conditional statements.
and: a and b (e.g., True and True results in True, True and False results in False).
or: a or b (e.g., True or False results in True, False or False results in False).
not: not a (e.g., not True results in False).


4. Bitwise Operators: These operators perform operations on the individual bits of a number.
& (AND): a & b (e.g., 10 & 3 results in 2).
| (OR): a | b (e.g., 10 | 3 results in 11).
^ (XOR): a ^ b (e.g., 10 ^ 3 results in 9).
~ (NOT): ~a (e.g., ~10 results in -11).
<< (Left Shift): a << b (e.g., 10 << 2 results in 40).









Q5). Explain the concept of type casting in Python with examples?
Concept of Type Casting in Python (with Examples):
Type casting in Python refers to the process of converting the data type of a variable from one type to another. Python provides built-in functions to perform this conversion.

Types of Type Casting:
1. Implicit Type Casting (Automatic)
Python automatically converts one data type to another during an operation, if needed.
No loss of data or precision.


Example:"""

x = 5        # int
y = 2.0      # float
result = x + y
print(result)       # Output: 7.0 (float)
print(type(result)) # <class 'float'>




"""2. Explicit Type Casting (Manual)
You manually convert a variable using functions like int(), float(), str(), etc.
Common Functions:
int() – converts to integer
float() – con
str() – converts to string
list(), tuple(), set() – converts to list, tuple, set


Examples:"""

# String to Integer
num_str = "100"
num_int = int(num_str)
print(num_int + 10)  # Output: 110

# Float to Integer
f = 5.6
print(int(f))        # Output: 5

# Integer to String
age = 25
print("Age is " + str(age))  # Output: Age is 25



"""Why Type Casting is Important:

It ensures compatibility between different data types.

Prevents errors during operations (e.g., you can't add a string and an integer without casting."""






"""Q6)How do conditional statements work in Python? Illustrate with examples.?

Conditional statements in Python are used to make decisions in code. They let the program choose different paths of execution based on whether a condition is True or False.

Types of Conditional Statements:

1. if Statement
Executes a block of code if the condition is True."""


x = 10
if x > 5:
    print("x is greater than 5")


---

"""2. if...else Statement

If the condition is True, one block of code runs; if False, another block runs."""


x = 3
if x > 5:
    print("x is greater than 5")
else:
    print("x is 5 or less")



"""3. if...elif...else Statement

Used for multiple conditions. Python checks each condition in order."""


x = 0
if x > 0:
    print("Positive number")
elif x == 0:
    print("Zero")
else:
    print("Negative number")



"""How It Works:

Python evaluates the condition.

If the condition is True, it executes the corresponding block of code.

If not, it moves to the next condition (elif) or the else block."""




"""Example with User Input:"""

age = int(input("Enter your age: "))

if age >= 18:
    print("You are an adult.")
else:
    print("You are a minor.")








"""Q7)Describe the different types of loops in Python and their use cases with examples?

Loops in Python are used to execute a block of code multiple times. Python provides two main types of loops: for loops and while loops.



1. for Loop

Definition:

Used to iterate over a sequence (like a list, tuple, string, or a range).

Syntax:"""

for variable in sequence:
    # code block

"""Example:"""

fruits = ['apple', 'banana', 'cherry']
for fruit in fruits:
    print(fruit)

"""Use Cases:

Iterating through elements in a list or string

Repeating a task a specific number of times using range()"""




"""2. while Loop

Definition:

Executes a block of code as long as a condition is True.

Syntax:"""

while condition:
    # code block

#Example:

i = 1
while i <= 5:
    print(i)
    i += 1

"""Use Cases:

When the number of iterations is not known in advance

Repeating a task until a user input or condition changes"""




"""3. Loop Control Statements


Example:"""

for i in range(5):
    if i == 3:
        continue
    print(i)



```
